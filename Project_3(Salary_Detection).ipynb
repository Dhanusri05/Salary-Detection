#Salary Detection (Project 3)

**Problem statement:** A HR wants to decide on the salary package for the candidate with available data(Experience,interview score, Test score and Salary)

**Understanding the problem:** The client needs to predict the salary based on the given dataset

```python
#Python Libraries for Data Analysis
#Pandas for Data Cleaning and Analysis
```

```python
import pandas as pd
```

```python
df=pd.read_csv('/content/Hiring DS.csv')
df
```

```python
import math
interview_scoremedian = math.floor(df['interview_score'].median())
interview_scoremedian
```

```python
df['interview_score'] = df['interview_score'].fillna(interview_scoremedian)
df
```

```python
df['Experience'] = df['Experience'].fillna('zero')
df
```

#Build ML Model

```python
#sklearn(scikit-learn) is a ML Library used for regression,classification and clustering 
```

```python
from sklearn import linear_model
reg = linear_model.LinearRegression()
```

```python
#machine trainng - fit(x,y) 
#x- input variables
#y- output variable
```

```python
x=df.drop('salary',axis=1)
x
```

```python
y=df['salary']
y
```

```python
reg.fit(x,y)
```

```python
reg.predict([[5,6,7]])
```

```python
reg.score(x,y)
```
